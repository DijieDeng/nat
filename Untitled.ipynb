```python
import numpy as np

np.random.seed(1)
```

```python
N = 100
d = 2


a1 = 0.5
a2 = 0.5
w = 1
r1 = np.random.rand()
r2 = np.random.rand()
```

```python
r1,r2
```

```python
def f1(X):
    return sum([x**2 for x in X])

def f2(X):
    return 10*len(X) + sum([(x**2 - 10*np.cos(2*np.pi*x)) for x in X])

def getX(n,d):
    X = []
    for i in range(n):
        xs = np.random.uniform(-5.13, 5.13, d)
        X.append(xs)
    return X

def update_v(v,a1,a2,w,x):
    v = w*v + a1*r1*(p-x)+a2*r2*(g-x)
    return v
```

```python
f1_x = [f1(x) for x in X]
p1 = g1 = [x for x in X[f1_x.index(min(f1_x))]]
```

```python
f2_x = [f2(x) for x in X]
p2 = g2 = [x for x in X[f2_x.index(min(f2_x))]]
```

```python
T= 20
X = getX(N,d)
V = [[0]*d]*N
p1 = X[1]
g1 = X[1]

f1_p = f1(p1)
f1_g = f1(g1)

for t in range(T):
    print(t)
    for i in range(N):
        V[i] = update_v(V[i],a1,a2,w,X[i])
        X[i] = X[i]+V[i]
        for j in range(d):
            if X[i][j] < -5.12:
                X[i][j] = -5.12
            elif X[i][j] > 5.12:
                X[i][j] = 5.12
        f1_x = f1(X[i]) 
        if f1_x < f1_p:
            f1_p = f1_x
            p1 = X[i]
            if f1_x < f1_g:
                f1_g = f1_x
                g1 = X[i]
                print(g1,t)
    print(X)
    

    
                
print(g1,f1_g)
print(p1,f1_p)
```

```python
T= 30

p = X[1]
g = X[1]
f2_p = f2(p)
f2_g = f2(g)

for t in range(T):
    for i in range(N):
        V[i] = update_v(V[i],a1,a2,w,X[i])
        X[i] = X[i]+V[i]
        for j in range(d):
            if X[i][j] < -5.12:
                X[i][j] = -5.12
            elif X[i][j] > 5.12:
                X[i][j] = 5.12
        f2_x = f2(X[i]) 
        if f2_x < f2_p:
            f2_p = f2_x
            p = X[i]
            if f2_x < f2_g:
                f2_g = f2_x
                g = X[i]
                

print(X)           
print(g,f1_g)
print(p,f1_p)
```
